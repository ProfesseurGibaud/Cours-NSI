## II) Parcours d'un Arbre

Il existe diverses façons de parcourir les nœuds d'un arbre.

 * Le <strong>parcours en largeur</strong>: les noeuds sont parcourus étage par étage, de haut en bas et de gauche à droite.
 * Le <strong>parcours en profondeur</strong>: on explore complétement un sous-arbre avant de commencer l'exploration de l'autre. Il existe trois façons de faire:
   - <strong>Parcours préfixe</strong> ou préordre (NGD): on visite d'abord le nœud, puis son sous-arbre gauche, puis son sous-arbre droit.
   - <strong>Parcours infixe </strong> ou en ordre (GND): on visite d'abord le sous-arbre gauche, puis le nœud, puis le sous-arbre droit.
   - <strong>Parcours postfixe </strong> ou en postordre (GDN): on visite d'abord le sous-arbre gauche, puis le sous-arbre droit, et enfin le noeud.
